```python
import py_entitymatching as em
```

```python
import py_stringmatching as sm
```

```python
PathA = 'stage3/two_csv_files/yelp.csv'
```

```python
PathB = 'stage3/two_csv_files/zomato.csv'
```

```python
A = em.read_csv_metadata(PathA,key = 'ID')
```

```python
B = em.read_csv_metadata(PathB,key = 'ID')
```

```python
A
```

```python
B
```

```python
ab = em.AttrEquivalenceBlocker()
C0 = ab.block_tables(A,B, 'Zipcode','Zipcode',l_output_attrs= ['Name','Address'] ,r_output_attrs= ['Name','Address'])
```

```python
C0
```

```python
ab = em.AttrEquivalenceBlocker()
C1 = ab.block_tables(A,B,'State','State',l_output_attrs= ['Name','Address'] ,r_output_attrs= ['Name','Address'])
```

```python
C1
```

```python
ft = em.get_features_for_blocking(A,B)
```

```python
ft[ft.left_attribute=='Zipcode']
```

```python
rb = em.RuleBasedBlocker()
rule = rb.add_rule(['Zipcode_Zipcode_lev_dist(ltuple,rtuple) > 1'],ft)
```

```python
C2 = rb.block_tables(A,B,l_output_attrs=['Name', 'Address'], r_output_attrs= ['Name', 'Address'])
```

```python
C2
```

```python
rb = em.RuleBasedBlocker()
rule = rb.add_rule([ 'Phone_Phone_lev_dist(ltuple,rtuple) > 1'],ft)
```

```python
C3= rb.block_tables(A,B,l_output_attrs=['Name','Address'],r_output_attrs=['Name', 'Address'])
```

```python
C3
```

```python
ab = em.AttrEquivalenceBlocker()
C4 = ab.block_tables(A,B, 'Address','Address',l_output_attrs= ['Name','Address'] ,r_output_attrs= ['Name','Address'])
```

```python
C4
```

```python
ab = em.AttrEquivalenceBlocker()
C5 = ab.block_tables(A,B, 'Name','Name',l_output_attrs= ['Name','Address'] ,r_output_attrs= ['Name','Address'])
```
